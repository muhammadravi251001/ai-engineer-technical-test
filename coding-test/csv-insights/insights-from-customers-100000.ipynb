**Main Question**: Please parse this csv, get the insight of the data `customer-100000.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
```

```python
FILE_PATH = "./data/customers-100000.csv"
```

```python
df = pd.read_csv(FILE_PATH)
df['Subscription Date'] = pd.to_datetime(df['Subscription Date'], errors='coerce') # Convert Subscription Date
df
```

```python
print("Dataset Info:")
df.info()
```

```python
print("Dataset Head:")
df.head()
```

```python
print("Unique values per column:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
top_cities = df['City'].value_counts().head(10)
print("Top 10 Cities:")
print(top_cities)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title("Top 10 Cities with Most Customers")
plt.xlabel("Number of Customers")
plt.ylabel("City")
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10)
print("Top 10 Countries:")
print(top_countries)

plt.figure(figsize=(8,4))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title("Top 10 Countries with Most Customers")
plt.xlabel("Number of Customers")
plt.ylabel("Country")
plt.show()
```

```python
top_companies = df['Company'].value_counts().head(10)
print("Top 10 Companies:")
print(top_companies)

plt.figure(figsize=(10,5))
sns.barplot(x=top_companies.values, y=top_companies.index, palette='coolwarm')
plt.title("Top 10 Companies with Most Customers")
plt.xlabel("Number of Customers")
plt.ylabel("Company")
plt.show()
```

```python
subscription_per_year = df.groupby(df['Subscription Date'].dt.year).size()
subscription_per_year.index = subscription_per_year.index.astype(str)
print("Subscription Per Year:")
print(subscription_per_year)

plt.figure(figsize=(10,4))
subscription_per_year.plot(kind='line', marker='o')
plt.title("Customer Subscription Trend per Year")
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.grid(True)
plt.show()
```

```python
subscription_per_month = df.groupby(df['Subscription Date'].dt.month).size()
subscription_per_month.index = subscription_per_month.index.map(lambda x: calendar.month_name[x])
print("Subscription Per Month:")
print(subscription_per_month)

plt.figure(figsize=(10,4))
subscription_per_month.plot(kind='line', marker='o')
plt.title("Customer Subscription Trend per Month")
plt.xlabel("Month")
plt.ylabel("Number of Customers")
plt.xticks(ticks=range(len(subscription_per_month.index)), labels=subscription_per_month.index, rotation=90)
plt.grid(True)
plt.show()
```
